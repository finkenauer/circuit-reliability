# LH
* Rising progatation delay - **in(a)**: pLHa

```python
df['pLHa'].describe()
```

```python
df.pLHa.sort_values().plot(kind='hist')
```

```python
import seaborn as sns

df.pLHa.plot(kind='hist')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, squeeze=False)

df.pLHa.plot(kind='hist', bins=50, ax=axes[0,0])
df.pLHa.plot(kind='hist', bins=100, ax=axes[0,1])
plt.grid()
```

```python
sns.kdeplot(df['pLHa'])
```

```python
sns.kdeplot(df['pLHa'], bw=50), sns.kdeplot(df['pLHa'], bw=100)
```

```python
plt.boxplot(df.pLHa)
```
