```python
%matplotlib inline
import pandas as pd
```

```python
dfA = pd.read_csv("monteCarlo_nand.measure", delim_whitespace=True, header=1)
dfA.head()
```

```python
dfB = pd.read_csv("monteCarlo_nandB.measure", delim_whitespace=True, header=1)
dfB.head()
```

```python
print(len(dfA), len(dfB))
```

```python
dfA.drop('Monte_Carlo_index', axis=1, inplace=True)
dfB.drop('Monte_Carlo_index', axis=1, inplace=True)
dfB.head()
```

# LH

```python
# Analisando a entrada B em fall
dfB.bpLH.describe()
```

```python
import seaborn as sns

dfA.apLH.plot(kind='hist')
```

```python
dfB.bpLH.plot(kind='hist')
plt.grid()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, squeeze=False)

dfA.apLH.sort_values().plot(kind='hist', ax=axes[0,0])
dfB.bpLH.sort_values().plot(kind='hist', ax=axes[0,1])
plt.grid()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, squeeze=False)

sns.kdeplot(dfA['apLH'], ax=axes[0,0])
sns.kdeplot(dfB.bpLH, ax=axes[0,1])
plt.grid()
```

```python
sns.kdeplot(dfA.apLH)
```

```python
plt.boxplot(dfA.apLH)
```

```python
plt.boxplot(dfB.bpLH)
```

```python
import scipy.stats as stats
print('Curtose - entrada A: ', stats.kurtosis(dfA.apLH, bias=False, fisher=False))
print('Curtose - entrada B: ', stats.kurtosis(dfB.bpLH, bias=False, fisher=False))
```

```python
print('Skewness - A: ', stats.skew(dfA.apLH, bias=False))
print('Skewness - B: ', stats.skew(dfB.bpLH, bias=False))
```

```python
import numpy as np 
import pylab 

measurements = dfA.apLH   
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
stats.probplot(dfB.bpLH, dist="norm", plot=pylab)
pylab.show()
```
